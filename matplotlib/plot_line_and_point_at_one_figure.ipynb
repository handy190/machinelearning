```python
line  = [1,3,2,5,4,7,6,9,8,11]
p = [5,5,5]
p_index = [0,4,6]
```

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize'] = (50.0, 10.0)


plt.xlabel('x')
plt.ylabel('y')
plt.title('close value')

plt.plot(range(len(line)), line, 'r')
plt.plot(range(len(p_index)), p, 'ob') # point shape (o,*,+), color blue


plt.show()
```
