```python
# 原始数据：职业-人数 分布
count_by_occupation = \
[(u'administrator', 79),
 (u'retired', 14),
 (u'lawyer', 12),
 (u'none', 9),
 (u'student', 196),
 (u'technician', 27),
 (u'programmer', 66),
 (u'salesman', 12),
 (u'homemaker', 7),
 (u'executive', 32),
 (u'doctor', 7),
 (u'entertainment', 18),
 (u'marketing', 26),
 (u'writer', 45),
 (u'scientist', 31),
 (u'educator', 95),
 (u'healthcare', 16),
 (u'librarian', 51),
 (u'artist', 28),
 (u'other', 105),
 (u'engineer', 67)]
```

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
x_occupation = np.array([c[0] for c in count_by_occupation])# 获取所有职业
y_count = np.array([c[1] for c in count_by_occupation])# 获取所有职业的从业人数分布

x = x_occupation[np.argsort(y_count)]# 将职业按照从业人数分布升序排序
y = y_count[np.argsort(y_count)]# 从业人数分布升序排序

width = 1
pos = np.arange(len(x))

#打印x轴的label
ax = plt.axes()
ax.set_xticks(pos+width/2)
ax.set_xticklabels(x)

#画bar图，设置x轴label斜向显示
plt.bar(pos, y, width)
plt.xticks(rotation=30)

#设置画布大小
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(16,10)
```
