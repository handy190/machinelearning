# 1. load data

```python
# load data
from sklearn.datasets import load_digits

digits = load_digits()

x_data = digits.data
y_data = digits.target
```

```python
x_data.shape
```

```python
y_data.shape
```

# 2. PCA

```python
# pca
from sklearn.decomposition import PCA
pca = PCA(n_components=3) # 3 for 3d plot
x_data_new = pca.fit_transform(x_data)
```

```python
x_data_new.shape
```

# 3. plot 3d

```python
# plot 3d
import joblib as jl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

new_X = x_data_new
y = y_data

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(new_X[:, 0], new_X[:, 1], new_X[:, 2], c=y, cmap=plt.cm.spectral)
plt.show()
```
