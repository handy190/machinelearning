# plot multi lines with labels

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(1)
y_true_train = [1,2,3,1,4,5]
y_true_valid = [2,3,7,4,3]
y_true_test = [1,2,3,7,9,6,3,2,1]

y_pred_train = [1.2,2.3,3.5,1.1,4.1,5.6]
y_pred_valid = [2,3,9,4,2]
y_pred_test = [1.8,2.5,3,6,10,7,2,1,1.5]

plt.plot(np.concatenate([y_true_train, y_true_valid, y_true_test]))
plt.plot(range(len(y_pred_train)), y_pred_train)
plt.plot(range(len(y_pred_train), len(y_pred_train+y_pred_valid)), y_pred_valid)
plt.plot(range(len(y_pred_train+y_pred_valid), len(y_pred_train+y_pred_valid+y_pred_test)), y_pred_test)
plt.ylabel('value')
plt.xlabel('time')
plt.title('result')
plt.legend(['truth', 'predict_train', 'predict_valid', 'predict_test'], loc='upper left')
plt.show()
```
