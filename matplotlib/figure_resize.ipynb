```python
from sklearn import datasets

digits = datasets.load_digits()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm


plt.imshow(digits.images[0], cmap=cm.binary)
plt.grid(False)
plt.axis('off')
plt.show()
```
