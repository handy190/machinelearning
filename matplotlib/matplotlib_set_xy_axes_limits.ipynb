# 1. if not set x/y axes limits

```python
import matplotlib.pyplot as plt

x = [i for i in range(10)]
y = [i*i for i in range(10)]

plt.plot(x, y)

plt.show()
```

# 2. if set x/y axes limits

```python
import matplotlib.pyplot as plt

x = [i for i in range(10)]
y = [i*i for i in range(10)]

plt.xlim( min(x)*2, max(x)*2)
plt.ylim( min(y)*3-50, max(y)*3)

plt.plot(x, y)
plt.show()
```
