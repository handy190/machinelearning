# 1. 2D Convolution

**averaging**
* computing the average of the pixel values inside that window
* convolution is a basic operation on image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE )# IMREAD_GRAYSCALE, IMREAD_COLOR

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img, cmap = 'gray', interpolation = 'bicubic'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst, cmap = 'gray', interpolation = 'bicubic'),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

# 2. Image Blurring

* blur = convolution with diff kernels

```python
blur = cv2.blur(img,(5,5))

plt.subplot(121),plt.imshow(img, cmap = 'gray', interpolation = 'bicubic'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur, cmap = 'gray', interpolation = 'bicubic'),plt.title('Blurring')
plt.xticks([]), plt.yticks([])
plt.show()
```

# 3. Gaussian Filtering

```python
blur = cv2.GaussianBlur(img,(5,5),0)
plt.subplot(121),plt.imshow(img, cmap = 'gray', interpolation = 'bicubic'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur, cmap = 'gray', interpolation = 'bicubic'),plt.title('Gaussian')
plt.xticks([]), plt.yticks([])
plt.show()
```

# 4. Median Filtering

```python
median = cv2.medianBlur(img,5)
plt.subplot(121),plt.imshow(img, cmap = 'gray', interpolation = 'bicubic'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median, cmap = 'gray', interpolation = 'bicubic'),plt.title('Median')
plt.xticks([]), plt.yticks([])
plt.show()
```

# 5. Bilateral Filtering

* highly effective at noise removal while preserving edges

```python
blur = cv2.bilateralFilter(img,9,75,75)
plt.subplot(121),plt.imshow(img, cmap = 'gray', interpolation = 'bicubic'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur, cmap = 'gray', interpolation = 'bicubic'),plt.title('Bilateral')
plt.xticks([]), plt.yticks([])
plt.show()
```
