# Perspective Transformation

* 透视变换
* 可以倾斜图像校正

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import cv2
import numpy as np

img = cv2.imread('img_src_rotated.png')
rows,cols,ch = img.shape

pts1 = np.float32([[63,12],[285,116],[15,115],[239,207]])
pts2 = np.float32([[0,0],[250,0],[0,100],[250,100]])

M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(250,100))

plt.figure(figsize=(20,10))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

# 如何设置参数

* pts1是原图（input）坐标，具体如下图中所示
* pts2是变换后的output图像大小，这里设置为长250，高100
