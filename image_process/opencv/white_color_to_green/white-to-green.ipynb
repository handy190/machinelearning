# modify all white color to green

```python
import cv2

# load image
p = 'Image20181128162821.jpg'
frame = cv2.imread(p, cv2.IMREAD_COLOR)
```

```python
%matplotlib inline
# display original image
from matplotlib import pyplot as plt
im = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(im, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# modify white color to green
img_tmp = frame.copy()

lower =(120, 120, 120) # lower bound for each channel
upper = (255, 255, 255) # upper bound for each channel

# create the mask and use it to change the colors
mask = cv2.inRange(img_tmp, lower, upper)
img_tmp[mask != 0] = [0,255,0]
```

```python
%matplotlib inline
# display modified image
from matplotlib import pyplot as plt
im = cv2.cvtColor(img_tmp, cv2.COLOR_BGR2RGB)
plt.imshow(im, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
