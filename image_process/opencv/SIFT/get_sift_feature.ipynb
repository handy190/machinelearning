# Get SIFT Feature

```python
import cv2
import numpy as np
```

```python
img_path = 'dataset/0012_3_1.bmp'
img = cv2.imread(img_path)
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sift = cv2.xfeatures2d.SIFT_create()
kp = sift.detect(gray,None) # get the SIFT features
print('type:{0}, len:{1}'.format(type(kp), len(kp)))
```

```python
# check the SIFT feature
f = kp[0]
for x in dir(f):
    if('__' not in x):
        print(x) # feature parameters
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

img=cv2.drawKeypoints(gray,kp,img) # draw the sift features

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

img=cv2.drawKeypoints(gray,kp,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
