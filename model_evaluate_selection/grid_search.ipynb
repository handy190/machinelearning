# GridSearch

http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html#sklearn.model_selection.GridSearchCV


```python
from sklearn.datasets import load_iris
iris = load_iris()
x_train, y_train = iris.data, iris.target# load data
```

```python
%%time
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVR
parameters = {'epsilon':[0,0.1,0.2,0.3,0.5,0.8], 'C':[1,2,3,5,8,10], 'max_iter':[100,1000,5000,10000], 'random_state':[88]}
model = LinearSVR()
model_gs = GridSearchCV(model, parameters)
model_gs.fit(x_train, y_train)
```

# Check selected model details

```python
# seledted best model
model_selected = model_gs.best_estimator_
```

```python
# seledted best model parameters
model_selected
```

```python
# best score
model_gs.best_score_ 
```
