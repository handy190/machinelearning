## 1. KNN usage

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))

#生成测试数据
n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# n_clusters:类簇的个数，即要将数据分成多少类
y_pred = KMeans(n_clusters=10, random_state=random_state).fit_predict(X)
```

## 2. KNN cluster graph

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))

#生成测试数据
n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# n_clusters:类簇的个数，即要将数据分成多少类
y_pred = KMeans(n_clusters=10, random_state=random_state).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)# X是二维的，X[:, 0]是图中的x轴, X[:, 1]是图中的y轴
plt.title("graph")
```
