# 1. SIR 模型的模拟

```python
%matplotlib inline

import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import odeint 
def SIR(sir, t, beta, gamma):
    "SIR模型的微分方程"
    S, I, R = sir 
    dsdt = - beta * S * I 
    didt = beta * S * I - gamma * I 
    drdt = gamma * I 
    return [dsdt, didt, drdt]
# 定义时间
t = np.linspace(0, 20, 5000) 
# 定义初始情况，易感染人数1000人，感染人1，恢复人0 
S0, I0, R0 = 1000, 1, 0 
# 感染比例与恢复比分别为
beta, gamma = 0.002, 0.25
# 求解时序变化
result = odeint(SIR, [S0, I0, R0], t, args=(beta, gamma))
St, It, Rt = result[:, 0], result[:, 1], result[:, 2] 
# 绘图
plt.plot(t, St, c="g", label="S") 
plt.plot(t, It, c="r", label="I") 
plt.plot(t, Rt, c="b", label="R") 
plt.title("SIR-Model", fontsize=18) 
plt.xlabel("Days", fontsize=18) 
plt.ylabel("Num", fontsize=18)
plt.legend(fontsize=18)
plt.grid(True)
plt.show()
```

# 2. SI模型研究

```python
%matplotlib inline

import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import odeint 
def SI(s, t, beta):
    "SI模型的微分方程"
    S, I = s
    dsdt = - beta * S * I 
    didt = beta * S * I 
    return [dsdt, didt]
# 定义时间
t = np.linspace(0, 20, 20) 
# 定义初始情况，易感染人数1000人，感染人1，恢复人0 
S0, I0 = 1000, 5
# 感染比例与恢复比分别为
beta = 0.001
# 求解时序变化
result = odeint(SI, [S0, I0], t, args=(beta,))
St, It = result[:, 0], result[:, 1]
# 绘图
plt.plot(t, St, c="g", label="S") 
plt.plot(t, It, c="r", label="I") 
plt.title("SI-Model", fontsize=18) 
plt.xlabel("Days", fontsize=18) 
plt.ylabel("Num", fontsize=18)
plt.legend(fontsize=18)
plt.grid(True)
plt.show()
```
