## 2. Training

```python
import word2vec
```

### 2.1 把文本变为词组

```python
%%time

word2vec.word2phrase('text8', 'text8-phrases')
```

下面查看结果

文件text8-phrases中，会将词组以aaa_bbb的形式显示出来

```python
%%time

txt = ''

with open('text8-phrases') as f:
    txt = f.readline()

word_list = txt.split(' ')
phase_list = []
for w in word_list:
    if('_' in w):
        phase_list.append(w)
```

```python
phase_list[:10] # 结果
```

### 2.2 把词组变为向量(模型)

```python
%%time

word2vec.word2vec('text8-phrases', 'text8.bin')
```

text8.bin是word2vec的模型，无法可视化查看

### 2.3 相似词聚类

```python
%%time

word2vec.word2clusters('text8', 'text8-clusters.txt', 100, verbose=True)
```

查看结果

text8-clusters.txt

```python
! vim text8-clusters.txt
```

说明： two, eight, five, three属于同一类（第39类）

## 3. Predict

### 3.1 加载模型

```python
%%time

model = word2vec.load('text8.bin')
```

### 3.2 查看模型向量

```python
# 查看维度
model.vectors.shape
```

```python
# 查看某个向量
model.vectors[0]
```

```python
# 查看某个词的向量
model['dog']
```

### 3.3 相似词语检索

以cosine相似度进行检索

```python
indexes, metrics = model.cosine('dog')
```

```python
# 相似词的index
indexes
```

```python
# 相似度
metrics
```

**通过下标查看具体的相似词组**

```python
model.vocab[indexes]
```

### 3.4 相似短语检索

```python
indexes, metrics = model.cosine('los_angeles')
model.generate_response(indexes, metrics).tolist()
# 输出与los_angeles相似的词组，及其位置
```

### 3.5 复杂推理

king - man + woman = queen

王的女人，并且不是男的，就是女王

```python
indexes, metrics = model.analogy(pos=['king', 'woman'], neg=['man'], n=10)
model.generate_response(indexes, metrics).tolist()
```
