```python
# auto play the game

from pklearn import Table
from pklearn.templates import simulate, BasicPlayer
from sklearn.ensemble import GradientBoostingRegressor


t = Table(smallBlind=1, bigBlind=2, maxBuyIn=200)

for i in range(6):
    
    r = GradientBoostingRegressor()
    name = 'Player ' + str(i+1)
    p = BasicPlayer(name=name, reg=r, bankroll=10**6, nRaises=10, rFactor=.7, memory=10**5)
    t.addPlayer(p)

simulate(t, nHands=10000, firstTrain=2000, nTrain=1000, nBuyIn=10)
simulate(t, nHands=20, nBuyIn=10, vocal=True)
```

## 2.2 create player, then play the game, then we can get feature

```python
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np
from pklearn import Table
from pklearn.templates import simulate, BasicPlayer
from sklearn.ensemble import GradientBoostingRegressor

# create player, then play the game
from pklearn import Table
from pklearn.templates import simulate, BasicPlayer
from sklearn.ensemble import GradientBoostingRegressor


t = Table(smallBlind=1, bigBlind=2, maxBuyIn=200)
player_list = []
for i in range(10):
    
    r = GradientBoostingRegressor()
    name = 'Player ' + str(i+1)
    p = BasicPlayer(name=name, reg=r, bankroll=10**6, nRaises=10, rFactor=.7, memory=10**5)
    t.addPlayer(p)
    player_list.append(p)

simulate(t, nHands=1000, firstTrain=200, nTrain=1000, nBuyIn=10)
```

```python
player = player_list[0]
features = np.array(player.getFeatures())
features.shape
```
