```python
import os


content = []
with open('README.md', 'r') as fr:
    for line in fr:
        content.append( line )

rps_src_dst = []
for f in os.listdir('.'):
    if '_' in f:
        src = f
        dst = '_'.join( src.split('_')[1:] )
        rps_src_dst.append( (src,dst) )
        print(src,dst)
        
```

```python
content
```

```python
result = []

for s in content:
    ss = s
    for x,y in rps_src_dst:
        if x in ss:
            orig = ss
            dst = ss.replace(x, y)
            ss = dst
            print(x,y,orig,dst)
    result.append(ss)
```
